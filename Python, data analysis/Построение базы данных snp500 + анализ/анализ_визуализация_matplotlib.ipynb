# 4.1.1. Анализ положений, связанных с инвестированием согласно теории Г.Марковица

## Распределение доходностей (дневных)

```python
N = 150 # число бинов гистограммы
for name in name_company:
    df1 = Quotes.loc[name].sort_values(by = 'Date')['Close'].pct_change(axis = 'rows') # обычные доходности
    df2 = Quotes.loc[name].sort_values(by = 'Date')["Close"].apply(np.log).diff() # логарифмические доходности
    x = np.linspace(min(df2.min(), df1.min()), max(df2.max(), df1.max()), N)
    plt.figure(figsize=(10, 6))
    plt.hist(df1, density=True, bins=N, label="Доходности", alpha=0.5)
    plt.hist(df2, density=True, bins=N, label="Логарифмические доходности", alpha=0.5)
    print(f"Коэф. ассиметрии доходностей = {df1.skew()}, Коэф. Куртозиса - 3 = {df1.kurt()-3}")
    print(f"Шапиро тест на нормальность доходностей {stats.shapiro(np.array(df1)[1:])}")
    print("-"*100)
    print(f"Коэф. ассиметрии логарифмических доходностей = {df2.skew()}, Коэф. Куртозиса - 3 = {df2.kurt()-3}")
    print(f"Шапиро тест на нормальность логарифмических доходностей: {stats.shapiro(np.array(df2)[1:])}")
    plt.plot(x, stats.norm.pdf(x, df2.mean(), df2.std()))
    plt.grid()
    plt.title(f"Гистограмма доходности компании: {name} (тикер)")
    plt.xlabel('Доходность')
    plt.show()
    plt.show()
```

Вывод - визуально распределения доходностей случайных 10 компаний отличаются от нормального распределения. Имеются выбросы. В качестве оценки, были для каждого распределения доходностей посчитаны коэф. ассиметрии и куртозиса. Коэф. куртозиса существенно больше, чем у нормального распределения, это обьясняет длинные хвосты и появление выбросов. В основном коэф. ассиметрии меньше нуля, то есть распределение скошено вправо, оно не симметрично, как нормальное. Был проведен Тест Шапиро на нормальность распределния, везде получилось очень маленькое p значение, значит 0 гипотезу отклоняем при любом "разумном" уровне значимости. Была построена гистограмма логарифмических доходностей, которая в некоторых случаях ближе к нормальному распределению, так как позволяет сгладить выбросы. Однако величины самих доходностей близки к нулю, следовательно логарифмические доходности практически равны обычным.

## Стабильность риска доходностей (дневных доходностей, усредненных по месяцам)

```python
Quotes1=Quotes.reset_index()
Quotes1['Date'] = pd.to_datetime(Quotes1['Date'])
Quotes1 = Quotes1.pivot(index='Date', columns='Ticker', values='Close')
Quotes1 = Quotes1.pct_change(axis = 'rows') # считаем доходности
Quotes1['year'], Quotes1['month'] = Quotes1.index.year, Quotes1.index.month
```

```python
Quotes2 = Quotes1.groupby(['year','month']).transform('std').drop_duplicates()
```

```python
for name in name_company:
    plt.figure(figsize=(10, 6))
    Quotes2[name].plot()
    plt.grid()
    plt.title(f"Динамика риска котировок компании {name} (тикер) по месяцам")
    plt.ylabel('Риск')
    plt.xlabel('Год')
    plt.show()
```

Средний риск для всех компаний волатилен на всем отрезке времени, он колеблется в коридоре порядка своего значения для большинства акций. У некоторых акций в пределах 2-3 значений. В 2020-2021 году наблюдается резкий скачек волатильности для всех компаний.

## Стабильность корреляций доходностей акций (корреляций дневных доходностей, усредненных по месяцам)

```python
def getCorrVar(ticker1,ticker2):
    def my_fun(x):
        return(x[ticker1].corr(x[ticker2]))
    X = Quotes1.groupby(['year','month']).apply(my_fun).drop_duplicates().reset_index()
    X["Date"] = X["year"].apply(lambda x: str(x)) +"-"+ X["month"].apply(lambda x: str(x))
    X['Date'] = pd.to_datetime(X['Date'])
    X = X[["Date", 0]].set_index("Date")
    f = lambda x, y: np.full(shape=len(X.index), fill_value=y, dtype=float)
    fig = plt.figure(figsize=(10, 6))
    axes = fig.add_axes([0,0,1,1])
    axes.plot(X)
    axes.plot(X.index, f(X, 0), label="Линия 0")
    axes.plot(X.index, f(X, X[0].mean()), label="Линия средней корреляции")
    axes.legend()
    plt.grid()
    plt.title(f"Динамика корреляции доходностей двух компаний: {ticker1} и {ticker2}")
    plt.ylabel('Значение корреляции')
    plt.show()
```

```python
for i in range(len(name_company)-1):
    getCorrVar(name_company[i], name_company[i+1])
```

Корреляции между выбранными компаниями в среднем положительна. Однако все корреляции доходностей совершенно нестабильны, в зависимости от времени они принимают положительные и отрицательные значения, наблюдаются резкие скачки волатильности. У большинства исследуемых компаний средняя корреляция составляет 0.3-0.4.

# 4.1.2. Анализ состояния финансового рынка

## Изменение средней доходности и риска по годам

```python
mean_year_profit = Quotes1.groupby("year").mean().stack() \
    .reset_index().set_index(["Ticker"]) \
    .rename(columns={0: "mean_profit"}).drop(["month"], axis=0)
```

```python
year_std = Quotes1.groupby("year").std().stack() \
    .reset_index().set_index(["Ticker"]) \
    .rename(columns={0: "std"}).drop(["month"], axis=0)
```

```python
year = pd.concat([mean_year_profit, year_std.drop(["year"], axis=1)], axis=1)
```

```python
year["mean_profit"] = year.mean_profit.apply(lambda x: (x+1)**252-1)
year["std"] = year["std"].apply(lambda x: np.sqrt(252)*x)
```

Таблица годовых доходностей и рисков для компаний snp500

```python
year.groupby("year").agg("mean")
```

## Что происходит на текущий момент времени с доходностью (например, последние полгода (рост, падение, колебание)) и уровень риска.

```python
Quotes5=Quotes.reset_index()
Quotes5['Date'] = pd.to_datetime(Quotes5['Date'])
Quotes5 = Quotes5.pivot(index='Date', columns='Ticker', values='Close')
```

```python
f = lambda x: x /x[0] -1
dataT1 = Quotes5.loc["2020-06-30":,].iloc[:, :-2].apply(f, axis=0)
```

```python
dataT1
```

```python
month_mean = Quotes1.drop(["year"], axis=1).loc["2020-07-01":,].groupby("month").mean().stack() \
    .reset_index().set_index(["Ticker"]) \
    .rename(columns={0: "mean_profit"})
month_std = Quotes1.drop(["year"], axis=1).loc["2020-07-01":,].groupby("month").std().stack() \
    .reset_index().set_index(["Ticker"]) \
    .rename(columns={0: "std"})
month = pd.concat([month_mean, month_std.drop(["month"], axis=1)], axis=1)
month["mean_profit"] = month.mean_profit.apply(lambda x: (x+1)**22-1)
month["std"] = month["std"].apply(lambda x: np.sqrt(22)*x)
```
